```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm', 'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )','12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
# Observe that From_To column shows data for From and To Destination values. This can be split into 2 columns using the '_' delimeter
```

```python
j = 0
df['From'] = ''
df['To'] = ''
while j != len(df):
    x = df['From_To'][j].split('_')
    df['From'][j] = x[0].capitalize()
    df['To'][j] = x[1].capitalize()
    j+=1
df
```

```python
df.drop(['From_To'], axis = 1)
```

```python
# Observe that in the FlightNumber column, 2 values are missing. These numbers are having a difference of 10 in between each other. 
# Hence the Flight number values will be 10055 and 10075.
```

```python
df = df.fillna(0)
df['FlightNumber'] = df['FlightNumber'].astype(int)
df
```

```python
i = 0
while i != len(df['FlightNumber']):
    if df['FlightNumber'][i] == 0:
        df['FlightNumber'][i] = df['FlightNumber'][i-1] + 10
    i = i+1
```

```python
# Dropping From_To column from our data frame
df = df.drop('From_To', axis=1)
```

```python
df
```

```python
# Expand the RecentDelays columns to multiple columns and renaming it with delays_1,delays_2, delays_3. Dropping the RecentDelays column at the end
```

```python
delays = df['RecentDelays'].apply(pd.Series)
delays.columns = ['delay_{}'.format(n) for n in range(1, len(delays.columns)+1)]
df = df.drop('RecentDelays', axis=1).join(delays)
df
```
